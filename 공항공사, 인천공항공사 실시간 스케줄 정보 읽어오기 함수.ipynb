```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from datetime import datetime

def select_list(url,params,table_class):
    head = {
       'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.84 Safari/537.36'
    }
    res = requests.post(url,headers=head,params=params)
    soup = bs(res.text,'lxml')
    tr_list = soup.select('.'+table_class+' tr')
    td_list = [list(map(lambda x:x.text.strip(),td)) for td in [tr.select('td')[1:] for tr in tr_list[1:]]]
    return td_list
```

```python
def kac_sch_list(port='GMP'):
    ## 한국공항공사 데이터
    params = {
        'airPort':port,
        'stHour':'01',
        'stMinute':'00',
        'edHour':'24',
        'edMinute':'55',
        'airType':'',
        'airLine':'ZE',
        'airLineNum':''
    }
    kac_list = []
    ## 한국공항공사 출발
    url = 'https://www.airport.co.kr/gimpo/extra/liveSchedule/liveScheduleList/layOut.do?langType=1&inoutType=OUT&cid=2015102611043202364&menuId=8'
    kac_list.extend([['D',d[0][2:],d[1],d[2],d[5],d[6]] for d in select_list(url,params,'table-responsive')])
    ## 한국공항공사 도착
    url = 'https://www.airport.co.kr/gimpo/extra/liveSchedule/liveScheduleList/layOut.do?langType=1&inoutType=IN&cid=2015102611052578964&menuId=10'
    kac_list.extend([['A',d[0][2:],d[1],d[2],d[5],d[6]] for d in select_list(url,params,'table-responsive')])
    df = pd.DataFrame(kac_list,columns=['type','flt','from','to','tm','gate'])
    df.to_csv(port+'_'+datetime.today().strftime('%Y%m%d')+'.csv')
    return df
```

```python
def icn_sch_list():
    today = datetime.today().strftime('%Y%m%d')
    ## 인천공항공사 데이터
    params = {
        'A':'A',
        'FROM_TIME':today+'0000',
        'TO_TIME':today+'2400',
        'AIRLINE':'ZE'
    }
    icn_list = []
    ## 인천공항공사 출발
    url = 'http://www.airport.kr/pa/aps/pac111000List.do'
    icn_list.extend([['D',d[0][2:],d[1],d[2],d[4],d[6]]
                     for d in select_list(url,params,'tablelist2')])
    ## 인천공항공사 도착
    url = 'http://www.airport.kr/pa/aps/pac112000List.do'
    icn_list.extend([['A',d[0][2:],d[1],d[2],d[4],d[5]]
                     for d in select_list(url,params,'tablelist2')])
    df = pd.DataFrame(icn_list,columns=['type','flt','from','to','tm','gate'])
    df.to_csv('ICN'+datetime.today().strftime('%Y%m%d')+'.csv')
    return df
```
